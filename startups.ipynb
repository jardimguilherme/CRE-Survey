```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = 'data.csv'
data = pd.read_csv(file)

data.head()
```

```python
# Visualização dos Anos de Experiência
plt.figure(figsize=(12, 8))
sns.histplot(data["Anos de Experiência"], bins=10, kde=True, color="skyblue")
plt.title("Distribuição dos Anos de Experiência")
plt.xlabel("Anos de Experiência")
plt.ylabel("Frequência")
plt.tight_layout()
plt.savefig('plots/audience/distribuicao_anos_experiencia.png', transparent=True)
plt.show()

# Visualização do Tamanho Médio da Equipe
plt.figure(figsize=(12, 8))
sns.histplot(data["Tamanho Médio da Equipe"], bins=10, kde=True, color="lightgreen")
plt.title("Distribuição do Tamanho Médio da Equipe")
plt.xlabel("Tamanho Médio da Equipe")
plt.ylabel("Frequência")
plt.tight_layout()
plt.savefig('plots/audience/distribuicao_tamanho_equipe.png', transparent=True)
plt.show()

# Visualização do Tempo de Mercado da Startup
plt.figure(figsize=(12, 8))
sns.histplot(data["Tempo de Mercado da Startup"], bins=10, kde=True, color="salmon")
plt.title("Distribuição do Tempo de Mercado da Startup")
plt.xlabel("Tempo de Mercado da Startup (anos)")
plt.ylabel("Frequência")
plt.tight_layout()
plt.savefig('plots/audience/distribuicao_tempo_mercado.png', transparent=True)
plt.show()

# Visualização de Cargo
plt.figure(figsize=(12, 8))
sns.countplot(y="Cargo", data=data, palette="Pastel1")
plt.title("Distribuição de Cargos")
plt.xlabel("Frequência")
plt.ylabel("Cargo")
plt.tight_layout()
plt.savefig('plots/audience/distribuicao_cargos.png', transparent=True)
plt.show()

# Gráfico de dispersão para Anos de Experiência vs Tamanho Médio da Equipe
plt.figure(figsize=(12, 8))
sns.scatterplot(x="Anos de Experiência", y="Tamanho Médio da Equipe", hue="Cargo", data=data, palette="Pastel1", s=100)
plt.title("Anos de Experiência vs Tamanho Médio da Equipe")
plt.xlabel("Anos de Experiência")
plt.ylabel("Tamanho Médio da Equipe")
plt.legend(title="Cargo", bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=6)
plt.tight_layout()
plt.savefig('plots/audience/scatter_anos_experiencia_tamanho_equipe.png', transparent=True, bbox_inches='tight')
plt.show()

# Gráfico de dispersão para Anos de Experiência vs Tempo de Mercado da Startup
plt.figure(figsize=(12, 8))
sns.scatterplot(x="Anos de Experiência", y="Tempo de Mercado da Startup", hue="Cargo", data=data, palette="Pastel1", s=100)
plt.title("Anos de Experiência vs Tempo de Mercado da Startup")
plt.xlabel("Anos de Experiência")
plt.ylabel("Tempo de Mercado da Startup (anos)")
plt.legend(title="Cargo", bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=6)
plt.tight_layout()
plt.savefig('plots/audience/scatter_anos_experiencia_tempo_mercado.png', transparent=True, bbox_inches='tight')
plt.show()
```
